```python
import openpyxl
import tkinter as tk
from tkinter import filedialog

def merge_excel_files(input_files, output_file):
    merged_workbook = openpyxl.Workbook()

    for file_path in input_files:
        workbook = openpyxl.load_workbook(file_path)
        for sheet_name in workbook.sheetnames:
            sheet = workbook[sheet_name]
            merged_sheet = merged_workbook.create_sheet(title=sheet_name)
            for row in sheet.iter_rows(values_only=True):
                merged_sheet.append(row)

    merged_workbook.save(output_file)

def browse_files():
    input_files = filedialog.askopenfilenames(
        title="Select Excel Files", filetypes=(("Excel files", "*.xlsx"), ("all files", "*.*"))
    )
    if input_files:
        files_entry.delete("1.0", tk.END)  # Clear any previous file selections
        for file in input_files:
            files_entry.insert(tk.END, file + "\n")

def merge_files():
    input_files = files_entry.get("1.0", tk.END).splitlines()
    output_file = filedialog.asksaveasfilename(
        defaultextension=".xlsx", filetypes=(("Excel files", "*.xlsx"), ("all files", "*.*"))
    )
    if input_files and output_file:
        merge_excel_files(input_files, output_file)
        result_label.config(text="Merge completed!")
    else:
        result_label.config(text="Merge canceled or no files selected.")

# Create the main window
root = tk.Tk()
root.title("Excel Merger")
root.geometry("600x400")  # Increase the window size

# Create and place widgets
label = tk.Label(root, text="Select Excel files to merge:")
label.pack()

files_entry = tk.Text(root, height=10, width=40)
files_entry.pack()

browse_button = tk.Button(root, text="Browse", command=browse_files)
browse_button.pack()

merge_button = tk.Button(root, text="Merge", command=merge_files)
merge_button.pack()

result_label = tk.Label(root, text="")
result_label.pack()

root.mainloop()
```
